```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Set style to 'white'
sns.set_style('white')

# Load titanic dataset into 'titanic', tips into 'tips' and iris into 'iris'
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
#You should mention the color name.
custom_clr = {'Male': 'gray', 'Female': 'green'}
sns.boxplot(x="day", y="tip", data=tips,hue = 'sex', palette= custom_clr)
```

```python
custom_clr = {'Male': 'Mulberry', 'Female': 'green'}
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex',palette= custom_clr)
plt.text(30, 9, 'pearsonr = 0.68; p = 6.7e-34',fontsize=8, color='black')
```

```python
sns.distplot(titanic['fare'] ,kde=False,bins=30)
```

```python
sns.boxplot(x="class", y="age", data=titanic,
           hue = 'alive')
```

```python

sns.jointplot(x='fare',y='age',data=titanic )
plt.text(300, 80, 'pearsonr = 0.096; p = 0.01',fontsize=8, color='black')
```

```python
sns.boxplot(x="class", y="age", data=titanic)
```

```python
sns.countplot(x="sex", hue="class",data=titanic)
```

```python
tc = titanic.corr()
sns.heatmap(tc,cmap='coolwarm',annot=True )
plt.title('titanic.corr()')
```
