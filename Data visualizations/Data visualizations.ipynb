```python
import matplotlib.pyplot as plt
import numpy as np

# Get the c and d values from the DataFrame
c = df['c']
d = df['d']

# Plot the scatter plot
plt.scatter(c, d, c='green', s=10)

# Set the title and labels of the axes
plt.title('c vs d')
plt.xlabel('c')
plt.ylabel('d')

# Set the size of the figure
plt.figure(figsize=(5, 5))

# Show the plot
plt.show()
```

```python
d = df['d']

plt.hist(d)

plt.title('Distribution of d')
plt.xlabel('d')
plt.ylabel('Frequency')

plt.show()
```

```python
d = df['d']

# Set the style sheet to bmh
plt.style.use('bmh')

# Create the histogram with 50 bins
plt.hist(d, bins=50)

# Set the title and labels of the axes
plt.title('Distribution of d')
plt.xlabel('d')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(2)
plt.figure(figsize=(4.5, 3.4))
df = pd.DataFrame(np.random.rand(1000, 4), columns=['a', 'b', 'c', 'd'])

# size of the data point based on col d
plt.scatter(df['a'], df['b'], s=df['d']*50, c=df['d'], cmap='coolwarm', alpha=1)

plt.xlabel('a')
plt.ylabel('b')

plt.colorbar(label='d')

# Create a boxplot for all columns in the DataFrame
plt.boxplot(df, patch_artist=True, color=dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray'), sym='r+')

# Set the title and labels of the axes
plt.title('Boxplot of all columns')
plt.xlabel('Column')
plt.ylabel('Value')

# Show the plot
plt.show()
```

```python

np.random.seed(2)

plt.figure(figsize=(5,3.4))
df = pd.DataFrame(np.random.rand(1000, 4), columns=['a', 'b', 'c', 'd'])

colors = dict(boxes='white', whiskers='DarkOrange', medians='DarkBlue', caps='Gray')
symbols = 'r+'

plt.boxplot(df.values, sym=symbols, vert=True, patch_artist=True, widths=0.5, notch=False, 
            boxprops=dict(facecolor=colors['boxes'], color='green'), 
            whiskerprops=dict(color=colors['whiskers']), 
            medianprops=dict(color=colors['medians'], linewidth=2), 
            capprops=dict(color=colors['caps'], linewidth=2))

plt.xticks(np.arange(1, len(df.columns) + 1), df.columns)

plt.show()
```
