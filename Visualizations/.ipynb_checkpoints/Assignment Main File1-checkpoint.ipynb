```python
import numpy as np
x = np.arange(0,10) 
y = x*2
z = x**2
print(x, y, z)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()

Fx = fig.add_axes([0, 0, 1, 1])
Fx.plot(x, y, label="y = x*2")
Fx.plot(x, z, label="y = x**2")
Fx.set_xlabel("x")
Fx.set_ylabel("y")
Fx.set_title("y vs. x")
plt.legend()
plt.show()
```

```python
fig = plt.figure()
Fx1 = fig.add_axes([0, 0, 1, 1])
Fx2 = fig.add_axes([0.1, 0.6, 0.2, 0.2])
Fx1.plot(x, z)
Fx2.plot(x, z)
Fx2.set_xlim([1, 2])
Fx2.set_ylim([0, 5])
Fx1.set_xlabel("x")
Fx1.set_ylabel("z")
Fx1.set_title("z vs. x")
plt.show()
```

```python
fig, Fxs = plt.subplots(1, 2, sharey=False)
for i in range(2):
    Fxs[i].plot(x, y)
    Fxs[i].set_xlim([1, 2])
    Fxs[i].set_ylim([0, 5])

for Fx in Fxs:
    Fx.set_xlabel("X")
    Fx.set_ylabel("Y")
plt.show()
```
