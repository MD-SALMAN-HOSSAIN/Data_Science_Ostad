```python
import numpy as np
x = np.arange(0,10) 
y = x*2
z = x**2
print(x, y, z)
```

```python
import matplotlib.pyplot as plt
```

```python
# Create the figure object
fig = plt.figure()

# Add an axis to the figure
ax = fig.add_axes([0, 0, 1, 1])

# Plot the data x, y on the axes
ax.plot(x, y, label="y = x*2")
ax.plot(x, z, label="y = x**2")

# Set the labels x and y
ax.set_xlabel("x")
ax.set_ylabel("y")

# Set the title
ax.set_title("y vs. x")

# Create legends
plt.legend()

# Show the plot
plt.show()
```

```python
# Create the figure object
fig = plt.figure()

# Add two axes to the figure
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.1, 0.6, 0.2, 0.2])

# Plot the data x, z on the main axes
ax1.plot(x, z)

# Plot the data x, z on the small axes and set its x limit b/w 1, 2, y limit 0, 5
ax2.plot(x, z)
ax2.set_xlim([1, 2])
ax2.set_ylim([0, 5])

# Replicate the figure to axes, titles, legends etc
ax1.set_xlabel("x")
ax1.set_ylabel("z")
ax1.set_title("z vs. x")
plt.legend()

# Show the plot
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, sharey=True)

# Plot the data x, y on the axes
for i in range(2):
    axs[i].plot(x, y)
    axs[i].set_xlim([1, 2])
    axs[i].set_ylim([0, 5])

# Replicate the figure to axes, titles, legends etc
for ax in axs:
    ax.set_xlabel("x")
    ax.set_ylabel("y")

plt.show()
```

```python
# Create the figure object
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the data x, y on the axes
axs[0].plot(x, y)
axs[1].plot(x, y)
axs[1].set_xlim([1, 2])
axs[1].set_ylim([0, 5])

# Replicate the figure to axes, titles, legends etc
for ax in axs:
    ax.set_xlabel("x")
    ax.set_ylabel("y")

plt.show()
```

```python
# Create the figure object
fig = plt.figure()

# Add three axes to the figure
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.1, 0.6, 0.2, 0.2])
ax3 = fig.add_axes([0.3, 0.2, 0.2, 0.2])

# Plot the data x, y on the first axis
ax1.plot(x, y, label="y = x*2")

# Plot the data x, z on the second axis
ax2.plot(x, z, label="y = x**2")

# Plot the data x, y on the third axis
ax3.plot(x, y, label="y = x*2")

# Set the labels x and y
ax1.set_xlabel("x")
ax1.set_ylabel("y")

# Set the title
ax1.set_title("y vs. x")

# Create legends
plt.legend()

# Show the plot
plt.show()
```
